```python
from scipy.stats import gaussian_kde
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x=np.random.normal(1,10,10000)
y=np.random.normal(1,50,10000)
arr = np.vstack([x, y])
z = gaussian_kde(arr)(arr)
idx = z.argsort()  
x, y, z = x[idx], y[idx], z[idx]
z /= z.max()   
plt.scatter(x, y, c=z, s=120, cmap = 'viridis', edgecolor = '', norm =mpl.colors.Normalize(vmin=0, vmax=z.max()))
#ax.scatter(x, y, c=z, s=120, cmap = 'viridis', edgecolor = ' ', norm = mpl.colors.Normalize(vmin=-0.2, vmax=z.max()))
```
