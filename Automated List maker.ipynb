```python
''' This is a program that makes a csv that has RA-Dec and observations
    It requires a list of just the RA-Dec (from iraf) and the oberservations
    Needs the files to be in the same folder    
'''
#this will take all of the files in your work directory and combine them
for root, dirs, filenames in os.walk('C:/Users/Erick/Desktop/work'):
    #iterates over all the files in your directory
    for i in range(len(filenames)):
        #finds which files have '_' which in my case are the files that have the ra and dec in them
        if '_' in filenames[i]:
            #This is where the single file program is generalized
            with open(root+'/'+filenames[i]) as xh:
                with open(root+'/'+filenames[i][5:13]+' - '+filenames[i][-9:]) as yh:
                    xlines = xh.readlines()
                    ylines = yh.readlines()
                    #b will have will have each row as an element of the array
                    b=[]
                    #takes the ra and dec columns and appends the observations to them
                    for j in range(len(xlines)):
                        a=[float(xlines[j].split()[0])]
                        a.append(float(xlines[j].split()[1]))
                        for k in range(8):
                            a.append(float(ylines[j].split()[k+2]))
                        #attaches the finished row a to the array b
                        b.append(a)
                    #writes each file rows as each of the rows of b
                    with open(root+'/'+filenames[i][5:13]+filenames[i][-9:-4]+'final.csv','w') as fi:
                        writer =csv.writer(fi,delimiter=',')
                        writer.writerows(b)
```
