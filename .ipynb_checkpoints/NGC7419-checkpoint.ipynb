```python
'''The goal of this program is to make a dictionary out of the data files
    The stucture of this dictionary should be {date:filter:[obser]}'''
for root, dirs, filenames in os.walk('C:/Users/Erick/Desktop/OU REU/NGC7419/Observations'):
    data = {}
    for fi in filenames:
        
        with open(root+'/'+fi) as f:
            date = fi[:8]
            fil = fi[11:12]
            a=[]
            test = csv.reader(f,delimiter='\t')
            for row in test:
                if row != []:
                    a.append(row)
            final = np.asarray(a)
            col= ['RA','Dec','q','qerr','u','uerr']
            print('for',fi,'the date is',date, ',filter is', fil)
            if date not in data.keys():
                print('whoops')
                data[str(date)]={fil:[final[:,0],final[:,1],final[:,2],final[:,4],final[:,4],final[:,5]]}
            else:
                data[str(date)][fil]=[final[:,0],final[:,1],final[:,2],final[:,4],final[:,4],final[:,5]]
        print(data)
```

```python
i=1
qu={}
for key in data.keys():
    plt.figure(i,figsize=(15,5),frameon=True)
    plt.suptitle(key,fontsize=40)
    n=1
    qu[key]={'B':[],'V':[],'I':[],'R':[]}
    for color in ['B','V','I','R']:
        plt.subplot(1,4,n)
        plt.grid()
        plt.title(color,fontsize=25)
        if color in data[key].keys():
            try:
                ind = cuts(key,color)
                plt.scatter(data[key][color][2].astype(float),data[key][color][4].astype(float),c='r',s=5,label=color)
                plt.scatter(data[key][color][2][(ind).astype(int)].astype(float),data[key][color][4][(ind).astype(int)].astype(float),c='b',s=5,label=color[0])
                qu[key][color]=[np.median(data[key][color][2][(ind).astype(int)].astype(float)),np.median(data[key][color][4][(ind).astype(int)].astype(float))]
                plt.scatter(qu[key][color][0],qu[key][color][1],s=30,c='k')
            except:
                plt.scatter(data[key][color][2].astype(float),data[key][color][4].astype(float),c='k',s=5,label=color)
        else:
            plt.text(-.055,-.01,'NO DATA',fontsize=40)
        plt.axhline(c='k',lw=1)
        plt.axvline(c='k',lw=1)
        n=n+1
        plt.axis('square')
    plt.tight_layout()
    i=i+1

def cuts(key,fil):
    ind, paralax,poe = np.loadtxt('C:/Users/Erick/Desktop/OU REU/NGC7419/Crossmatch stars/'+fil.lower()+key+'-result.csv', skiprows=1,usecols=(97,10,12),unpack=True,delimiter=',')
    paralax=paralax[np.logical_and(paralax>0,poe>5)] #This is our quality cut
    test=ap.sigma_clip(paralax,sigma=3,iters=5)
    finalcut=ind[np.ma.nonzero(test)]-1
    return finalcut
```
